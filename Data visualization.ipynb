```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flower= sns.load_dataset("iris")
flower
```

```python
flower.species.unique()
```

```python
sns.set_style("darkgrid")
```

```python
plt.plot(flower.sepal_length, flower.sepal_width);
```

```python
sns.scatterplot(flower.sepal_length, flower.sepal_width);
```

```python
sns.scatterplot(flower.sepal_length, flower.sepal_width, hue=flower.species);
```

```python
plt.figure(figsize=(12,6))
plt.title("Sepal Dimension")
sns.scatterplot(flower.sepal_length, flower.sepal_width, hue=flower.species,s=80);
```

```python
plt.title("Sepal Width")
plt.hist(flower.sepal_width);
```

```python
plt.hist(flower.sepal_width, bins=5);
```

```python
plt.hist(flower.sepal_width, bins=15);
```

```python
import numpy as np
plt.hist(flower.sepal_width, bins=np.arange(2,5, 0.25));
```

```python
setosa= flower[flower.species=='setosa']
versicolor= flower[flower.species=='versicolor']
virginica= flower[flower.species=='virginica']
```

```python
plt.hist(setosa.sepal_width, alpha = 0.5, bins=np.arange(2,5, 0.25));
plt.hist(versicolor.sepal_width, alpha = 0.5, bins=np.arange(2,5, 0.25));
plt.legend(['setosa','versicolor'])
```

```python
plt.hist([setosa.sepal_width,versicolor.sepal_width,virginica.sepal_width], alpha = 0.5, bins=np.arange(2,5, 0.25), stacked=True);
plt.legend(['setosa','versicolor','virginica'])
```

```python
plt.hist([setosa.sepal_width,versicolor.sepal_width,virginica.sepal_width], alpha = 0.5, bins=np.arange(2,5, 0.25));
plt.legend(['setosa','versicolor','virginica'])
```

```python
year=range(2000,2006)
apple=[0.35,0.6,0.9,0.8,0.65,0.8]
orange=[0.4,0.8,0.9,0.7,0.6,0.8]
```

```python
plt.plot(year,orange);
```

```python
plt.bar(year,orange);
```

```python
plt.plot(year,orange,'o--r');
plt.bar(year,orange);
plt.title("Yield of Oranges")
```

```python
plt.bar(year,orange);
plt.bar(year,apple, bottom=apple);
plt.legend(['oranges','apples'])
```
